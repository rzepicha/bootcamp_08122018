```python
im2 = cv2.imread("images/index.jpeg")
sum_array=im2.sum(axis=2, keepdims=True)
sum_array=sum_array/np.amax(sum_array)

data=sum_array*np.ones(3)
data.shape

plt.imshow(data)
```
